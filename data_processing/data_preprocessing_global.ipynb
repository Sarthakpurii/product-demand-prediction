```python
for column in ['StoreID', 'ProductName', 'Promo_applied', 'specials', 'mods', 'weather']:
    print(column,' : ',df_train[column].unique())
    print()
```

```python
df_train['Promo_applied'] = df_train['Promo_applied'].map({
    '1': 1,
    'Yes': 1,
    'Disc': 1,
    np.nan: 0
}).fillna(0).astype(int)
```

```python
df_train['specials']=df_train['specials'].map({
    'Yes':1,
    'No':0,
    'Y':1,
    'N':0,
    'nO':0
})
```

```python
df_train['mods']=df_train['mods'].map({
    'Y':1,
    'No':0
})
```
