```python
df = df.sort_values('Day_Count')

cutoff = int(df.shape[0] * 0.8)
cutoff_day = df.iloc[cutoff]['Day_Count']

df_train = df[df['Day_Count'] < cutoff_day]
df_test = df[df['Day_Count'] >= cutoff_day]
```

```python
print("Cutoff day:", cutoff_day) 
```

```python
df_train.to_csv('train.csv', index=False)
df_test.to_csv('test.csv', index=False)
```

```python
df_train = pd.read_csv('train.csv')
```

```python
df_train.columns
```

```python
for column in ['StoreID', 'ProductName', 'Promo_applied', 'specials', 'mods', 'weather']:
    print(column,' : ',df_train[column].unique())
    print()
```

```python
df_train['Promo_applied'] = df_train['Promo_applied'].map({
    '1': 1,
    'Yes': 1,
    'Disc': 1,
    np.nan: 0
}).fillna(0).astype(int)
```

```python
df_train['specials']=df_train['specials'].map({
    'Yes':1,
    'No':0,
    'Y':1,
    'N':0,
    'nO':0
})
```

```python
df_train['mods']=df_train['mods'].map({
    'Y':1,
    'No':0
})
```
